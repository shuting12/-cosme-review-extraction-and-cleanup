```python
def clean_df(df):
    df['Review date']=pd.to_datetime(df['Review date'])
    df['month']=df['Review date'].dt.month
    df['year']=df['Review date'].dt.year
    df['week']=df['Review date'].dt.week
    df['quarter'] = df['Review date'].dt.quarter
    df['Rating_value']=df.Rating.str.extract('(\d+)').fillna(0).astype(int)
    df['Purchaser or repeater']=df.Rating.str.replace('\d+', '')
    return df

clean_df=clean_df(Reviews)
```

```python
def age_range(row):
    if row <20:
        return 'younger than 20'
    if (row>=20 and row<=25) :
        return '21-25'
    if (row>=26 and row<=30) :
        return '26-30'
    if (row>=31 and row<=40) :
        return '31-40'    
    if (row>=41 and row<=50) :
        return '41-50'
    if (row>=51 and row<=60) :
        return '51-60'

    else:        
        return 'Older than 60'
clean_df['Age_range']=clean_df.Age.apply(lambda row: age_range(row))
```

```python
# by age group
age_df=pd.DataFrame(clean_df.groupby(['Age_range','year','quarter','month'])['Rating_value'].mean())
age_df['num_reviews']=clean_df.groupby(['Age_range','year','quarter','month'])['Review date'].count()
age_df=age_df.reset_index()
age_df.head()
```

```python
def dfs_tabs(df_list, sheet_list, file_name):
    writer = pd.ExcelWriter(file_name,engine='xlsxwriter')   
    for dataframe, sheet in zip(df_list, sheet_list):
        dataframe.to_excel(writer, sheet_name=sheet, startrow=0 , startcol=0)   
    writer.save()

dfs = [clean_df,
       Daily_review,
       month_review,
       quarter_review,
       year_review]

sheets = ['Raw Data',
          'Daily_data',
          'Monthly_Data',
          'Quarterly_Data',
          'Yearly_Data']       

# dont forget to change the name of file
dfs_tabs(dfs, sheets, 'Lancome_japan_cosme_full_reviews.xlsx')        
```
