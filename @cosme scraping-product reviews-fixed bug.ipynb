```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import re

# import requests package
import requests                   
import urllib.request
import urllib.error
from lxml.html import etree
import pandas as pd
import numpy as np

from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry



# import BeautifulSoup from package bs4 (i.e. beautifulsoup4)

from bs4 import BeautifulSoup 
import nltk
from nltk.tokenize import RegexpTokenizer
import nagisa
#nltk.download()
```

```python
# contains many infomation need later process
def get_name(review):
    name=''
    customer_name=review.select("div.reviewer-info a")
    if customer_name!=[]:
        try:
            name=customer_name[0].get_text("\t").split()[0]
        except: ''
    return name

def get_auth(review):
    auth=''
    customer=review.select("div.reviewer-info a span.ico-cmn-auth")
    if auth!=[]:
        try:
            auth=customer[0].get_text()
        except: ''
    return auth

def get_rating(review):
    rating=''
    product_info=review.select("div[class='rating clearfix'] p")
    if rating!=[]:
        try:
            rating=product_info[0].get_text()
        except: ''
    return rating

def get_product_review_date(review):
    review_date=''
    product_info=review.select("div[class='rating clearfix']")
    if review_date!=[]:
        try:
            review_date=product_info[0].get_text().split()[1]
        except: 
            review_date='NA'
    return review_date



def get_reviewer_age(review):
    reviews_age=''
    customer=review.select("div.reviewer-info")
    if customer!=[]:
        try:
            reviews_age=customer[0].get_text().split()[-1]
            reviews_age=re.findall('\d+', reviews_age)[0]   # convert it to numbers
        except:''
    return reviews_age


def get_no_reviews(review):
    no_reviews=''
    customer=review.select("div.reviewer-info")
    if customer!=[]:
        try:
            no_reviews=customer[0].get_text().split()[-1]
            no_reviews=re.findall('\d+', no_reviews)[1]   # convert it to numbers
        except:''
    return no_reviews

def get_skin_type(review):
    skin_type=''
    customer=review.select("div.reviewer-info ul li:nth-child(2)")
    if customer!=[]:
        try:
            skin_type=customer[0].get_text().split()[-1]
        except:''
    return skin_type

# contains many infomation need later process
def get_key_words(review):
    keywords=''
    product_info=review.select("div[class='tag-list clearfix'] dl")
    if keywords!=[]:
        try:
            keywords=product_info[0].get_text().split()[2:]
        except: ''
    return keywords

# contains many infomation need later process
def get_purchased_location(review):
    location=''
    product_info=review.select("div[class='tag-list clearfix'] dl")
    if location!=[]:
        try:
            location=product_info[0].get_text().split()[1]
        except: ''
    return location
```

# Compile all functions for final web scrapper

```python
def cosme(item_code, page):
    
    authorization=[]
    name=[]    
    product_review_date=[]
    rating_value=[]
    reviewer_age=[]
    skin_type=[]
    num_reviews=[]
    key_words=[]
    purchased_location=[]

    
    for i in range(1, page):
        url='https://www.cosme.net/product/product_id/'+ item_code+ '/reviews/p/' + str(i)
        headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36', 'Referer': '{}'.format(url)}
        session=requests.Session()
# in case of error, retry at most 3 times, waiting
# at least half a second between each retry        
        retry = Retry(total=3, backoff_factor=0.5)  
        adapter = HTTPAdapter(max_retries=retry)
        session.mount('http://', adapter)
        session.mount('https://', adapter)
        
        
        url = session.get(url,headers=headers)    
        req = url.text.encode('utf-8').decode('UTF-8')
        #req = requests.get(url, headers=headers)
        reviews=[]
        if url.status_code==200: 
            soup = BeautifulSoup(url.content, 'html.parser')
            reviews=soup.select("div[class='review-sec']")




            
            
            
            for review in reviews:
                product_review_date.append(get_product_review_date(review)),
                name.append(get_name(review)),
                authorization.append(get_auth(review)),
                rating_value.append(get_rating(review)),
                reviewer_age.append(get_reviewer_age(review)),
                num_reviews.append(get_no_reviews(review)),
                skin_type.append(get_skin_type(review)),
                key_words.append(get_key_words(review)),
                purchased_location.append(get_purchased_location(review))
      
    
            review_list = pd.DataFrame(
                {'Review date':product_review_date,
                 'Name': name,
                 'Authorization': authorization,  
                 'Rating': rating_value,
                 'Age': reviewer_age,
                 'Skin_type': skin_type,
                 'Number of Reviews': num_reviews,
                 'Key Words': key_words,
                 'Purchased_location':purchased_location})
            
            review_list.to_csv(item_code + '_rating.csv', index=False, encoding='utf_8_sig')
            
            
            #print(review_list)
            
            
```
